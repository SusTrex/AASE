```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset
dataset = pd.read_csv("../AASE/datasets/apartments_for_rent_classified_100K.csv", 
                       encoding='ISO-8859-1', sep=';', on_bad_lines='skip', 
                       low_memory=False)
```

```python
# Grfico de barras para preo mdio por categoria
plt.figure(figsize=(5, 3))
sns.barplot(x='category', y='price', data=dataset, estimator=lambda x: sum(x) / len(x))
plt.title('Preo Mdio por Categoria')
plt.ylabel('Preo Mdio')
plt.xticks(rotation=45)
plt.show()
```

```python
# Grfico de caixa para preo em relao  categoria
plt.figure(figsize=(8, 5))
sns.boxplot(x='category', y='price', data=dataset)
plt.title('Distribuio do Preo por Categoria de Apartamento')
plt.xlabel('Categoria')
plt.ylabel('Preo')
plt.xticks(rotation=45)
plt.show()
```

```python
# Grfico de barras para preo mdio por cidade
plt.figure(figsize=(12, 6))
sns.barplot(x='cityname', y='price', data=dataset, estimator=lambda x: sum(x) / len(x))
plt.title('Preo Mdio por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Preo Mdio')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='state', y='price', data=dataset, estimator=lambda x: sum(x) / len(x))
plt.title('Preo Mdio por Estado')
plt.xlabel('Estado')
plt.ylabel('Preo Mdio')
plt.xticks(rotation=45)
plt.show()
```

```python
# Grfico de barras para preo mdio por tipo de pagamento
plt.figure(figsize=(8, 4))
sns.barplot(x='price_type', y='price', data=dataset, estimator=lambda x: sum(x) / len(x))
plt.title('Preo Mdio por Tipo de Pagamento')
plt.xlabel('Tipo de Pagamento')
plt.ylabel('Preo Mdio')
plt.show()
```

```python
# Grfico de caixa para preo em relao  categoria
plt.figure(figsize=(8, 4))
sns.boxplot(x='category', y='price', data=dataset)
plt.title('Distribuio do Preo por Categoria de Apartamento')
plt.ylabel('Categoria')
plt.xlabel('Preo')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='category', hue='has_photo', data=dataset)
plt.title('Presena de Foto por Categoria de Imvel')
plt.xlabel('Categoria')
plt.ylabel('Contagem')
plt.legend(title='Tem Foto')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='bedrooms', data=dataset)
plt.title('Distribuio de Quartos por Categoria de Imvel')
plt.xlabel('Categoria')
plt.ylabel('Nmero de Quartos')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='state', hue='fee', data=dataset)
plt.title('Distribuio de Imveis com Taxas por Estado')
plt.xlabel('Estado')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.legend(title='Tem Taxa')
plt.show()
```
