```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset
dataset = pd.read_csv("../AASE/datasets/apartments_for_rent_classified_100K.csv", 
                       encoding='ISO-8859-1', sep=';', on_bad_lines='skip', 
                       low_memory=False)
```

```python
# Gráfico de barras para preço médio por categoria
plt.figure(figsize=(5, 3))
sns.barplot(x='category', y='price', data=dataset, estimator=lambda x: sum(x) / len(x))
plt.title('Preço Médio por Categoria')
plt.ylabel('Preço Médio')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico de caixa para preço em relação à categoria
plt.figure(figsize=(8, 5))
sns.boxplot(x='category', y='price', data=dataset)
plt.title('Distribuição do Preço por Categoria de Apartamento')
plt.xlabel('Categoria')
plt.ylabel('Preço')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico de barras para preço médio por cidade
plt.figure(figsize=(12, 6))
sns.barplot(x='cityname', y='price', data=dataset, estimator=lambda x: sum(x) / len(x))
plt.title('Preço Médio por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Preço Médio')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='state', y='price', data=dataset, estimator=lambda x: sum(x) / len(x))
plt.title('Preço Médio por Estado')
plt.xlabel('Estado')
plt.ylabel('Preço Médio')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico de barras para preço médio por tipo de pagamento
plt.figure(figsize=(8, 4))
sns.barplot(x='price_type', y='price', data=dataset, estimator=lambda x: sum(x) / len(x))
plt.title('Preço Médio por Tipo de Pagamento')
plt.xlabel('Tipo de Pagamento')
plt.ylabel('Preço Médio')
plt.show()
```

```python
# Gráfico de caixa para preço em relação à categoria
plt.figure(figsize=(8, 4))
sns.boxplot(x='category', y='price', data=dataset)
plt.title('Distribuição do Preço por Categoria de Apartamento')
plt.ylabel('Categoria')
plt.xlabel('Preço')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='category', hue='has_photo', data=dataset)
plt.title('Presença de Foto por Categoria de Imóvel')
plt.xlabel('Categoria')
plt.ylabel('Contagem')
plt.legend(title='Tem Foto')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='bedrooms', data=dataset)
plt.title('Distribuição de Quartos por Categoria de Imóvel')
plt.xlabel('Categoria')
plt.ylabel('Número de Quartos')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='state', hue='fee', data=dataset)
plt.title('Distribuição de Imóveis com Taxas por Estado')
plt.xlabel('Estado')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.legend(title='Tem Taxa')
plt.show()
```
