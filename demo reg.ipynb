### Importar os módulos e funções necessários

```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn import tree
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import graphviz
from sklearn.metrics import mean_absolute_error
```

### Carregar o conjunto de dados

https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html#sklearn.datasets.load_diabetes

```python
diabetes = load_diabetes()
diabetes
```

```python
X = diabetes.data
y = diabetes.target
```

```python
pd.DataFrame(X, columns=diabetes.feature_names)
```

### Dividir os dados em treino (70%) e teste (30%)

```python
Xtr, Xts, ytr, yts = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Criar uma Árvore de Decisão e treiná-la no conjunto de dados de treino

```python
dt = tree.DecisionTreeRegressor(max_depth=2, random_state=42)
dt.fit(Xtr, ytr)
```

### Visualizar a Árvore de Decisão

```python
plt.figure(figsize=(20, 7))
tree.plot_tree(dt, feature_names=diabetes.feature_names)
plt.plot()
```

### Realizar previsões no conjunto de dados de teste

```python
prev = dt.predict(Xts)
prev
```

### Avaliar a qualidade das previsões

```python
# MAE - mean squared error
mean_absolute_error(yts, prev)
```

```python
# NMAE - normalized mean squared error (%)
(mean_absolute_error(yts, prev) / (yts.max() - yts.min())) * 100
```
